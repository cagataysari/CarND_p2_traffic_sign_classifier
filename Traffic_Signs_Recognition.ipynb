```python
### Define your architecture here.
### Feel free to use as many code cells as needed.
import tensorflow as tf

#Hyper-parameters
learning_rate = 0.02
epochs = 5
batch_size = 100
display_step  = 1  

# Output depth
k_output = 32


# Image Properties
image_width, image_height, color_channels = image_shape

# Convolution filter - the image size of 32x32 is small 
filter_size_width = 6
filter_size_height = 6
filter_stride = [1, 2, 2, 1] 

filter_element_num = int( (image_width/(filter_stride[1]))          \
                          *                                         \
                          (image_height/(filter_stride[2]))   \
                          *                                         \
                          k_output                                  \
                         )
            

#hidden layer
n_hidden_layer = 256 # layer number of features
```

```python
# Input/Image - [batch_size, height, width, channels] 
nn_input = tf.placeholder( 
                        tf.float32, 
                        shape=[None, image_width, image_height, color_channels]
                       )

nn_output = tf.placeholder(tf.float32, [None, n_classes])



# Store layers weight & bias
weight = {
    'convnet':tf.Variable(tf.truncated_normal( 
                         [filter_size_width, filter_size_height, color_channels, k_output])
                     ),
    
    'fc7_hidden_layer': tf.Variable(tf.random_normal([filter_element_num, n_hidden_layer])),
    'out': tf.Variable(tf.random_normal([n_hidden_layer, n_classes]))
}
bias = { # has thweighte number of the layer outputs
    'convnet': tf.Variable(tf.zeros(k_output)),
#     'convnet': tf.Variable(tf.constant(0.05, shape=[k_output])), 
    'fc7_hidden_layer': tf.Variable(tf.random_normal([n_hidden_layer])),
    'out': tf.Variable(tf.random_normal([n_classes]))
}

# Apply Convolution 
conv_layer_weight = tf.nn.conv2d(nn_input, weight['convnet'], filter_stride, padding='SAME') 
#Stide:  the input tensor has 4 dimensions: [batch, height, width, channels] 
 
# Add bias 
conv_layer_bias = tf.nn.bias_add(conv_layer_weight, bias['convnet']) 
# Apply activation function 
conv_layer = tf.nn.relu(conv_layer_bias)
```

```python
conv_layer

```

```python
conv_layer_flat = tf.reshape(conv_layer,[-1,filter_element_num])
```

```python
conv_layer_flat
```

```python
# Hidden layer with RELU activation
fc7_layer = tf.add(tf.matmul(conv_layer_flat, weight['fc7_hidden_layer']), bias['fc7_hidden_layer'])
fc7_layer = tf.nn.relu(fc7_layer)

# Output layer with linear activation
logits = tf.matmul(fc7_layer, weight['out']) + bias['out']


# Define loss and optimizer
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, nn_output))

# cost = tf.reduce_mean(-tf.reduce_sum(nn_output * tf.log(tf.clip_by_value(tf.nn.softmax(logits),1e-10,1.0)), reduction_indices=[1]))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)

# Initializing the variables
init = tf.initialize_all_variables()


correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(nn_output, 1))
# Calculate validation accuracy
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
```

```python
def findNextBatch(i, total_num):
    #todo: SGD -shuffle_batch_join - mnist.train.next_batch(batch_size)
    
    if (i == (total_num-1)):
        #last batch does not necessarily have enough examples
        return InputSet_Train[i*batch_size:], OutputLabel_Train[i*batch_size:]
    else:
        return InputSet_Train[i*batch_size:(i+1)*batch_size], OutputLabel_Train[i*batch_size:(i+1)*batch_size]




# Launch the graph
with tf.Session() as sess:
    
    sess.run(init)
    weight_conv = sess.run(weight['convnet'])
    print( "INITAL:  weight_conv_shape: ", weight_conv.shape)    
    print( "np.count_nonzero(weight_conv) :  ", np.count_nonzero(weight_conv) )
    weight_conv_nan_table = np.isnan(weight_conv)
    print( "weight conv - num of NaN :  ", np.count_nonzero(weight_conv_nan_table) )
    
    # Training cycle
    for epoch in range(epochs):
        total_batch = int(size_train/batch_size) 
        # Loop over all batches
        for i in range(total_batch):
            batch_x, batch_y = findNextBatch(i, total_batch)

            batch_y_one_hot = np.eye(n_classes)[batch_y]

            # Run optimization op (backprop) and cost op (to get loss value)
            sess.run(optimizer, feed_dict={nn_input: batch_x, nn_output: batch_y_one_hot})
            
#             weight_conv = sess.run(weight['convnet'])
#             weight_conv_nan_table = np.isnan(weight_conv)
#             print("batch index: ",i, " | Conv Weights - num of NaN:  ", np.count_nonzero(weight_conv_nan_table) )            

        # Display logs per epoch step
        if epoch % display_step == 0:
            c = sess.run(cost, feed_dict={nn_input: batch_x, nn_output: batch_y_one_hot})
            print("Epoch:", '%04d' % (epoch+1), "cost=", \
                "{:.9f}".format(c))
            
#             debug_accu_logits = (tf.argmax(logits, 1)).eval({nn_input: batch_x})
#             print("   logits_prediction:",debug_accu_logits)
#             debug_accu_label = (tf.argmax(nn_output, 1)).eval({nn_output: batch_y_one_hot})
#             print("   labels:",debug_accu_label)
#             print("   debug_accu_logits_shape:",logits.get_shape().as_list())

#             debug_accu_logits_val = (logits).eval({nn_input: batch_x})
#             print("   logits_val:",debug_accu_logits)
#             debug_accu_fc7_val = (fc7_layer).eval({nn_input: batch_x})
#             print("   fc7_layer:",debug_accu_fc7_val)
#             debug_accu_convn_flat = (conv_layer_flat).eval({nn_input: batch_x})
#             print("   conv_layer_flat:",debug_accu_convn_flat)    

#             debug_accu_convn_w = (tf.reduce_sum(conv_layer_weight)).eval({nn_input: batch_x})
#             print("   conv_layer_weight sum:",debug_accu_convn_w) 
        
#             debug_accu_convn_b = (tf.reduce_sum(conv_layer_bias)).eval({nn_input: batch_x})
#             print("   conv_layer_bias sum:",debug_accu_convn_b)          
#             debug_accu_convn = (tf.reduce_sum(conv_layer)).eval({nn_input: batch_x})
#             print("   debug_accu_convn sum:",debug_accu_convn)   
            
            weight_conv = sess.run(weight['convnet'])
            print( "LAST batch: weight_conv_shape: ", weight_conv.shape,"np.count_nonzero(weight_conv) :  ", np.count_nonzero(weight_conv) )
            weight_conv_nan_table = np.isnan(weight_conv)
            print( "weight conv - num of nan :  ", np.count_nonzero(weight_conv_nan_table) )
#             print("batch_x: ", batch_x)
#             debug_accu_conv_layer_flat = (conv_layer_flat).eval({nn_input: batch_x})
#             print("   debug_accu_conv_layer_flat:",debug_accu_conv_layer_flat)
                        
            
            accu_vali = accuracy.eval({nn_input: batch_x, nn_output: batch_y_one_hot})
            print("Accuracy(training batch):", accu_vali)

    print("Optimization Finished!")

    # Test model

    OutputLebel_Validation_one_hot = np.eye(n_classes)[OutputLabel_Validation]
    accu_vali = accuracy.eval({nn_input: InputSet_Validation, nn_output: OutputLebel_Validation_one_hot})
    print("Accuracy - Validation:", accu_vali)

    # Calculate test accuracy
    OutputLabel_Train_one_hot = np.eye(n_classes)[OutputLabel_Train]
    accu_test = accuracy.eval({nn_input: InputSet_Train, nn_output: OutputLabel_Train_one_hot})
    print("Accuracy - Train:", accu_test)

   
```
